```python
%matplotlib inline
import numpy as np
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd

# Define the instruments to download. We would like to see Apple, Microsoft and the S&P500 index.
tickers = ['SPY']

# We would like all available data from 01/01/2000 until 2019-12-31.
start_date = '2010-01-01'
end_date = '2019-12-31'

# User pandas_reader.data.DataReader to load the desired data. As simple as that.
df = data.DataReader(tickers, 'yahoo', start_date, end_date)

```

```python
df
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

X = df[['High', 'Low', 'Open', 'Volume']]
#print(X)
y = df['Close']
#print(y)
X[['High', 'Low', 'Open', 'Volume']] = scale.fit_transform(X[['High', 'Low', 'Open', 'Volume']].values)

print(X)
```

```python
est = sm.OLS(y, X).fit()
```

```python
est.summary()
```
